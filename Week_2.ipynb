
# 1. NumPy Basics

NumPy is used for numerical computing and efficient array operations.


```python

# Creating arrays

arr = np.array([10, 20, 30, 40, 50])

print(arr)
print("Shape:", arr.shape)
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
```

```python

# Matrix operations

matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)

print("Transpose:")
print(matrix.T)
```

```python
# 5 students × 3 subjects (Math, Science, English)
marks = np.array([
    [78, 82, 85],
    [91, 89, 94],
    [67, 72, 70],
    [88, 90, 86],
    [56, 60, 58]
])

student_names = ["Aarav", "Diya", "Rohan", "Ananya", "Kabir"]
subject_names = ["Math", "Science", "English"]

print("Marks Array:")
print(marks)
```

```python
print('Total marks of each student: ',np.sum(marks,axis=1)) #Marks of each row added as no of students = rows and no of subjects = column
print('Average marks of each student: ',np.mean(marks,axis=1,dtype='f')) #Mean of each row taken as int
```

```python
print('Maximum marks of each subject: ',np.max(marks,axis=0)) #Max of each column taken
print('Minimum marks of each subject: ',np.min(marks,axis=0)) #Min of each column taken
```


# 3. Matplotlib Basics

Matplotlib is used for plotting and visualizing data.


```python
# Note: Make sure you have completed the Pandas section above
# so that the Total_Marks column exists in the dataframe.
# Line plot of student averages
plt.figure(figsize=(8, 5))

plt.plot(df["Name"], df["Average"], marker='o')

plt.title("Student Average Scores")
plt.xlabel("Student Name")
plt.ylabel("Average Score")

plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python

# Bar chart

subject_means = [
    df["Math"].mean(),
    df["Science"].mean(),
    df["English"].mean()
]

subjects = ["Math", "Science", "English"]

plt.bar(subjects, subject_means)

plt.title("Average Marks by Subject")
plt.ylabel("Average Marks")

plt.show()
```

```python

# Histogram

all_marks = pd.concat([
    df["Math"],
    df["Science"],
    df["English"]
])

plt.hist(all_marks, bins=10)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
df['Total_Marks']=df[['Math','Science','English']].sum(axis=1)
df['Average_Marks']=df[['Math','Science','English']].mean(axis=1)
print(df.loc[df['Total_Marks'].idxmax()]) 
print('\n')
print(df.loc[df['Total_Marks'].idxmin()]) 
print('\n')
print(df.loc[df['Average_Marks']>80]) 
```

```python
lst=np.array(df[['Math','Science','English']])
Subjects=['Math','Science','English']
Mean=np.mean(lst)
Median=np.median(lst)
Standard_Deviation=np.std(lst)
print(Mean," ",Median," ",Standard_Deviation)
Sub_Mean=np.mean(lst, axis=0)
print(Subjects[np.argmax(Sub_Mean)])
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(df['Average_Marks'], marker='o')
plt.title("Student Averages")
plt.xlabel("Student")
plt.ylabel("Average")
plt.grid(True)

plt.subplot(2,2,2)
bins=np.arange(50,100,5)
plt.hist(lst, bins=bins, edgecolor='black')
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.subplot(2,2,3)
plt.bar(Subjects, Sub_Mean, color=['r', 'g', 'b'])
plt.title("Subject-wise Averages")
plt.ylabel("Average")
plt.show()
```
